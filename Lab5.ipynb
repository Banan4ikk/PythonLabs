Напишите код для построения двух или более линий с разными стилями

```python
import matplotlib.pyplot as plt
# Точки 1 линии
pointX1 = [0,28,32]
pointY1 = [0,48,21]
# Точки второй линии
pointX2 = [0,21,33]
pointY2 = [50,20,32]
# Устанавливаем метку для подписи оси Х
plt.xlabel('Ось Х')
# Устанавливаем метку для подписи оси Y
plt.ylabel('Ось Y')
# Устанавливаем стили для линий
plt.plot(pointX1,pointY1, color='black', linewidth = 3,  label ='Линия 1 - точечная',linestyle='dotted')
plt.plot(pointX2,pointY2, color='green', linewidth = 3,  label ='Линия 2 - пунктрирная', linestyle='dashed')
# Set a title 
plt.title("График для построения нескольких линий")
plt.legend()
plt.show()
```
